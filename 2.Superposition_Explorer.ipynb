1. Scaling Superposition ($n$ Qubits)
   
Applying a Hadamard gate to a single qubit in the $|0\rangle$ state creates a superposition with 2 possible measurement outcomes. When we scale this up and apply Hadamard gates to $n$ qubits initialized in the $|0\rangle$ state, we create a uniform superposition containing $2^n$ possible classical states. For a 2-qubit system ($n=2$), this yields $2^2 = 4$ states: $|00\rangle$, $|01\rangle$, $|10\rangle$, and $|11\rangle$.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

n = 2
qc = QuantumCircuit(n)


qc.h(range(n))

qc.draw('mpl')
```

2. Statevectors and Probability Amplitudes:

   In quantum mechanics, the probability of measuring a specific state is the absolute square of its probability amplitude ($Probability = |Amplitude|^2$). Because a uniform superposition gives every state an equal measurement probability of $\frac{1}{2^n}$, the mathematical amplitude for each state in our statevector is precisely $\frac{1}{\sqrt{2^n}}$. For our 2-qubit system, this means every state has an amplitude of $\frac{1}{2}$.

```python
state = Statevector(qc)
print("Mathematical Amplitudes:", state.data)


probabilities = state.probabilities_dict()
print("Measurement Probabilities:", probabilities)
plot_histogram(probabilities)
```

3. Visualizing with the Q-Sphere
   
The Q-Sphere provides a comprehensive 3D representation of the statevector before measurement collapses the system. The size of each node represents the probability of measuring that specific state, while the color of the node represents the quantum phase. Because the Hadamard operations involve strictly real numbers, the phase for all uniform states remains at 0 degrees.

```python
display(plot_state_qsphere(state))
```
